```python
import pandas as pd

# âœ… Use the exact filename and folder path
file_path = r"C:\Users\Admin\OneDrive\Documents\data science\CSV files\NHS_TESTING_DATA.CSV"
df = pd.read_csv(file_path)

# ðŸ“Œ Confirm successful load
print("Data loaded successfully. Here's a preview:")
print(df.head())
```

```python
numeric_columns = [
    'newTestsByPublishDate',
    'cumTestsByPublishDate',
    'plannedCapacityByPublishDate'
]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
print("Updated data types:\n", df[numeric_columns].dtypes)
print("\nMissing values after conversion:\n", df[numeric_columns].isna().sum())
```

```python
# Summary statistics
print(df[numeric_columns].describe())

# Visual peek
import matplotlib.pyplot as plt

df[numeric_columns].plot(kind='line', title='Testing Metrics Over Time')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
